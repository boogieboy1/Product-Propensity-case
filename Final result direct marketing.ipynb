```python
import numpy as np
import pandas as pd
```

# Read the results of the regression models

```python
data_CL= pd.read_csv("consumer_loan_prop_pred_value.csv")
data_CC = pd.read_csv("credit_card_prop_pred_value.csv")
data_ML = pd.read_csv("mutual_fund_prop_pred_value.csv")
```

# Combine the datasets

```python
data = data_CL.merge(data_CC, how='outer', on='Client')
data = data.merge(data_ML, how='outer', on='Client')
```

```python
data.replace(np.NaN, 0, inplace=True)
```

```python
data['revenue_sum']= data['Sale_CL_pred_value']+ data['Sale_CC_pred_value']+ data['Sale_MF_pred_value']
```

```python
simple_result = data[['Client','Sale_CL_pred_value','Sale_CC_pred_value', 'Sale_MF_pred_value','Sale_CL_pred','Sale_CC_pred','Sale_MF_pred','revenue_sum']]
```

```python
top100 = simple_result.sort_values(ascending = False, by='revenue_sum').head(100)
```
